```python
# Part 1
with open("full.txt") as f:
    data = f.read().splitlines()

data = [[k for k in d] for d in data]
count = 0
for i, line in enumerate(data):
    if i == 0:
        continue
    # print(line)
    for j, item in enumerate(line):
        if item == '.':
            if data[i-1][j] == '|' or data[i-1][j] == 'S':
                data[i][j] = '|'

        elif item == '^':
            if data[i-1][j] == '|':
                data[i][j-1] = '|'
                data[i][j+1] = '|'
                count += 1
print(count)
# for line in data:
#     print("".join(line))
```

```python
# Part 2
with open("full.txt") as f:
    data = f.read().splitlines()

data = [[k if k != '.' else 0 for k in d ] for d in data]

for i, line in enumerate(data):
    if i == 0:
        continue
    
    for j, item in enumerate(line):
        if type(item) == int:
            if data[i-1][j] == 'S':
                data[i][j] = 1
            if (type(data[i-1][j]) == int and data[i-1][j] > 0):
                data[i][j] += data[i-1][j]
        elif item == '^':
            if type(data[i-1][j]) == int and data[i-1][j] > 0:
                data[i][j-1] += data[i-1][j]
                data[i][j+1] += data[i-1][j]

print(sum(data[-1]))
# for line in data:
#     for entry in line:
#         print(entry, end=" ")
#     print()

```
