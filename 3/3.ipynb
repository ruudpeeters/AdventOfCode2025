```python
with open("full.txt") as f:
    data = f.read().splitlines()

def get_maximum_and_index(string: str):
    m = 0
    index = -1
    for ind, s in enumerate(string):
        i = int(s)
        if i > m:
            m = i
            index = ind

    if index == -1:
        raise ValueError(f"Didn't find maximum {m} in string {s}")

    return m, index

def get_joltage(string: str):
    size = 12
    values = []
    start_index = 0
    for entry in range(size):
        if size-entry-1 == 0:
            value, index = get_maximum_and_index(string[start_index:])
        else:
            value, index = get_maximum_and_index(string[start_index:-(size-entry-1)])
        values.append(value)
        start_index += index + 1

    joltage_string = ""
    for val in values:
        joltage_string += str(val)

    return int(joltage_string)
    

total_joltage = 0

for line in data:
    total_joltage += get_joltage(line)

print(total_joltage)
```
