```python
with open("test.txt") as f:
    data = f.read()

ranges, ids = data.split('\n\n')

fresh_ranges=[]
for r in ranges.splitlines():
    fresh_ranges.append([int(s) for s in r.split('-')])
```

```python
# Part 1
count = 0
for id in ids.splitlines():
    for fr in fresh_ranges:
        if int(id) in range(fr[0], fr[1] + 1):
            count += 1
            break
print(count)
```

```python
# Part 2
from typing import List

def merge_two_ranges(c, d):
    if c[0] < d[0]:
        return merge_two_ranges_ordered(c, d)
    else:
        return merge_two_ranges_ordered(d, c)

def merge_two_ranges_ordered(a: List,b: List):
    if not a[0] <= b[0]:
        raise ValueError(f"Expecting ordered lists, got {a}, {b}")
    if a[1] < b[0]:
        return[a,b]
    elif a[1] in range(b[0], b[1] + 1):
        return[[a[0], b[1]], []]
    elif a[1] > b[1] + 1:
        return [a,[]]

out_ranges = []
for r in fresh_ranges:
    print(out_ranges)
    # Add first entry
    if out_ranges == []:
        out_ranges.append(r)
        continue

    print("Checking ", r)
    # Determine entrypoint first value
    pos_start = len(out_ranges) + 1
    for i, out_r in enumerate(out_ranges):
        if r[0] < out_r[0] or r[0] in range(out_r[0], out_r[1] + 1):
            pos_start = i
            break

    # Determine entrypoint second value
    pos_end = len(out_ranges) + 1
    if pos_start == len(out_ranges) + 1:
        out_ranges.append(r)  # If new range start after the last one, don't need to check endpoint
        continue
    else:
        for j, out_r in enumerate(out_ranges):
            if r[1] < out_r[0] or r[1] in range(out_r[0], out_r[1] + 1):
                pos_end = j
                break
    
    # print(pos_start, pos_end)

    if pos_start > pos_end:
        raise ValueError(f"Found pos_start larger than pos_end: {pos_start}, {pos_end}")

    if pos_start == pos_end:
        # Get new ranges to put in output_ranges
        out_1, out_2 = merge_two_ranges(r, out_ranges[pos_start])
        
        # Put new ranges in proper location
        if out_2 == []:
            out_ranges[pos_start] = out_1
        else:
            out_ranges = out_ranges[:pos_start] + [out_1, out_2] + out_ranges[pos_start+1:]
    else:
        iter_count = 0
        for iter, k in enumerate(out_ranges[pos_start: pos_end + 1].copy()):
            print("iter", iter, k)
            out_1, out_2 = merge_two_ranges(r, k)
            print("1,2", out_1, out_2)
            if out_2 == []:
                out_ranges[pos_start + iter + iter_count] = out_1
                if pos_start + iter + iter_count + 1 == len(out_ranges):
                    continue
                out_3, out_4 = merge_two_ranges(out_ranges[pos_start + iter + iter_count], out_ranges[pos_start + iter + iter_count + 1])
                print("3,4", out_3, out_4)
                if out_4 == []:
                    iter_count -= 1
                    out_ranges[pos_start + iter + iter_count] = out_3
                    del out_ranges[pos_start + iter + iter_count + 1]
                else:
                    out_ranges[pos_start + iter + iter_count] = out_3
                    out_ranges[pos_start + iter + iter_count + 1] = out_4
            else:
                out_ranges = out_ranges[:pos_start + iter + iter_count] + [out_1, out_2] + out_ranges[pos_start + iter + iter_count +1:]
                iter_count += 1
            print("out_range", out_ranges)

print(out_ranges)
```
