```python
# Part 1
import numpy as np

def do_sum(a):
    if a[-1] == '*':
        tot = 1
        for i in a[:-1]:
            tot *= int(i)
        return tot
    if a[-1] == '+':
        tot = 0
        for i in a[:-1]:
            tot += int(i)
        return tot

with open("test.txt") as f:
    data = f.read().splitlines()

data = [d.split() for d in data]

arr = np.array(data).transpose([1,0])

total = 0
for a in arr:
    total += do_sum(a)
    
    
print(total)
```

```python
import numpy as np


def parse_sum(problem):
    trans = np.array(problem).transpose((0,1))
    operator = ""
    numbers = []
    print(trans)
    for line in trans:
        if line[-1] == '*' or line[-1] == '+':
            print(line[0])
            operator = line[-1]

        s = int("".join(line[:-1]))
        numbers.append(s)

    if operator == '*':
        tot = 1
        for n in numbers:
            tot *= n
        print("tot *", tot)
        return tot
    elif operator == '+':
        print('tot +', sum(numbers))
        return sum(numbers)

with open("full.txt") as f:
    data = f.read().splitlines()

data = [[i for i in d] for d in data]

arr = np.array(data).transpose()

sums = []
current_sum = []
for line in arr:
    if all(l == " " for l in line):
        sums.append(current_sum)
        current_sum = []
    else:
        current_sum.append(line)
sums.append(current_sum)

total = 0
for individual_sum in sums:
    total += parse_sum(individual_sum)
print(total)

```
