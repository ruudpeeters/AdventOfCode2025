```python
def parse_value(value: str, dict: dict, count: int, end_value: str) -> int:
    if value == end_value:
        return 1
    if value == "out":
        return 0
    
    total = 0
    for v in dict[value]:
        total += parse_value(v, dict, count, end_value)
    return total

with open("full.txt") as f:
    data = f.read().splitlines()

device_dict = {}
for line in data:
    key, values = line.split(':')
    if key in device_dict:
        raise ValueError(f"Key {key} already in device_dict")
    device_dict[key] = values.split()

print(parse_value('wjt', device_dict, 0, "dac"))
```
