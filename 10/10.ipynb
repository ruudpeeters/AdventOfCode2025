```python
# Part 1

from itertools import combinations_with_replacement

def parse_input(string):
    # Parse target
    target = string.split()[0]
    target = [1 if val == "#" else 0 for val in target[1:-1]]

    # Parse costs
    costs = string.split()[1:-1]
    costs = [eval(c) for c in costs]
    costs = [[c] if type(c) is int else list(c) for c in costs]
    
    # Parse joltage
    joltages = eval(string.split()[-1])

    return target, costs, joltages

with open("full.txt") as f:
    data = f.read().splitlines()

total = 0
for line in data:
    target, buttons, _ = parse_input(line)
    # print("Target:", target)
    # print("Buttons:", buttons)

    N = 1
    found = False
    while True:
        for button_combination in combinations_with_replacement(buttons, N):
            result = [0 for i in range(len(target))]
            for button in button_combination:
                for light in button:
                    result[light] = ~result[light] & 1

            if all([result[j] == target[j] for j in range(len(target))]):
                # print(result, target)
                found = True
                break
        if found:
            break
        N += 1
    total += N
print(total)
```

```python
# Part 2

from itertools import combinations_with_replacement

def parse_input(string):
    # Parse target
    target = string.split()[0]
    target = [1 if val == "#" else 0 for val in target[1:-1]]

    # Parse costs
    costs = string.split()[1:-1]
    costs = [eval(c) for c in costs]
    costs = [[c] if type(c) is int else list(c) for c in costs]
    
    # Parse joltage
    joltages = string.split()[-1]
    joltages = eval(f"[{joltages[1:-1]}]")

    return target, costs, joltages

with open("test.txt") as f:
    data = f.read().splitlines()

total = 0
for iter, line in enumerate(data):
    print(f"Parsing line {iter} of {len(data)}")
    _, buttons, joltages = parse_input(line)
    print("joltages:", joltages)
    print("Buttons:", buttons)

    line_count = 0
    count = 0
    while any([joltage != 0 for joltage in joltages]):
        # print()
        print(joltages)
        # Find joltage value to fix first
        non_zero_joltages = []
        for jolt in joltages:
            if jolt != 0:
                non_zero_joltages.append(jolt)
        minimum_joltage = min(non_zero_joltages)
        # print("minimum joltage", minimum_joltage)
        min_joltage_index = joltages.index(minimum_joltage)
        # print("min joltage index", min_joltage_index)

        # Find largest button containing this joltage value
        sorted_buttons = sorted(buttons, key= lambda x: len(x), reverse=True)
        # print(sorted_buttons)
        button = -1
        for sorted_button in sorted_buttons:
            # print(sorted_button, min_joltage_index in sorted_button)

            if min_joltage_index in sorted_button:
                valid = True
                for sb in sorted_button:
                    # print(sb, joltages[sb])
                    if joltages[sb] == 0:
                        valid = False
                        break
                if not valid:
                    continue
                button = sorted_button
                break
        if button == -1:
            raise ValueError("Didn't find a button")
        # print(button)
        for index in range(len(joltages)):
            if index in button:
                joltages[index] -= minimum_joltage
        # print(joltages)
        line_count += minimum_joltage
    # print("Line count", line_count)
    total += line_count

print("Total:", total)
```
